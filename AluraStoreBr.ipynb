### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento_loja_1 = loja['Preço'].sum()
faturamento_loja_2 = loja2['Preço'].sum()
faturamento_loja_3 = loja3['Preço'].sum()
faturamento_loja_4 = loja4['Preço'].sum()

print(f"Faturamento Total da Loja 1: R$ {faturamento_loja_1:.2f}")
print(f"Faturamento Total da Loja 2: R$ {faturamento_loja_2:.2f}")
print(f"Faturamento Total da Loja 3: R$ {faturamento_loja_3:.2f}")
print(f"Faturamento Total da Loja 4: R$ {faturamento_loja_4:.2f}")
```

```python
import matplotlib.pyplot as plt

# Certifique-se de que você já calculou essas variáveis:
# faturamento_loja_1 = loja['Preço'].sum()
# faturamento_loja_2 = loja2['Preço'].sum()
# faturamento_loja_3 = loja3['Preço'].sum()
# faturamento_loja_4 = loja4['Preço'].sum()

faturamentos = [faturamento_loja_1, faturamento_loja_2, faturamento_loja_3, faturamento_loja_4]
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(10, 6))
plt.bar(lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Relatório Final: Análise de Desempenho da Alura Store

### Introdução

Este relatório apresenta uma análise do desempenho das quatro lojas da rede de e-commerce Alura Store, com o objetivo de auxiliar o Sr. João a decidir qual delas vender para obter capital para um novo negócio. A decisão será baseada em uma análise de métricas financeiras, operacionais e de satisfação do cliente, com o intuito de identificar a loja com o menor desempenho geral.

---

### Metodologia de Análise

Para a tomada de decisão, foram avaliados os seguintes indicadores-chave de desempenho (KPIs) para cada loja:

1.  **Faturamento Total:** Soma de todas as vendas para avaliar a receita bruta.
2.  **Média de Avaliação:** Média das avaliações dos clientes para medir a satisfação e a qualidade do serviço.
3.  **Custo Médio do Frete:** Gasto médio por venda com frete, importante para a análise da margem de lucro.
4.  **Desempenho de Vendas:** Identificação das categorias e produtos mais e menos vendidos para entender o mix de vendas.

---

### Resultados da Análise

#### **1. Faturamento Total**

Com base na análise de faturamento, a distribuição da receita entre as lojas é a seguinte:

* **Loja 1:** R$ 1.534509,12

* **Loja 2:** R$ 1.488459,06

* **Loja 3:** R$ 1.464025,03

* **Loja 4:** R$ 1.384497,58

Como podemos ver no gráfico, a **Loja 1** se destaca com o maior faturamento, enquanto a **Loja 4** apresenta o menor faturamento entre as quatro.


*Observação: A Loja 1 possui uma receita significativamente maior, o que a torna a mais rentável. A Loja 4, por outro lado, é a de menor receita e, consequentemente, a primeira a ser considerada para venda.*

#### **2. Média de Avaliação dos Clientes**

A satisfação do cliente é um fator crucial. A média de avaliação de 1 a 5 para cada loja é:

* **Loja 1:** 3.98
* **Loja 2:** 4.04
* **Loja 3:** 4.05
* **Loja 4:** 4.00

O gráfico de avaliação demonstra que todas as lojas mantêm uma média de avaliação alta, próxima de 4.0, indicando um bom nível de satisfação geral. No entanto, a **Loja 3** tem a maior média, mostrando que, apesar do baixo faturamento, os clientes que compram nela estão satisfeitos.


#### **3. Custo Médio do Frete**

O custo médio de frete impacta diretamente a margem de lucro de cada venda. Os resultados foram:

* **Loja 1:** R$ 34,69

* **Loja 2:** R$ 33,62

* **Loja 3:** R$ 33,07

* **Loja 4:** R$ 31,28

Curiosamente, a *Loja 4, que tem o menor faturamento, também apresenta o **menor custo médio de frete. Isso sugere uma operação de logística mais eficiente em comparação às outras lojas. Por outro lado, a **Loja 1*, que lidera em faturamento, tem o maior custo de frete.

---

### *Conclusão e Recomendação*

Com base na análise das métricas, a *Loja 4* é a candidata ideal para ser vendida. Apesar de sua eficiência em frete, a loja possui o *faturamento total mais baixo* entre todas as quatro. Embora a operação seja mais eficiente do ponto de vista de custos logísticos, a receita gerada não é suficiente para torná-la competitiva em relação às outras filiais.

A *Loja 1*, por exemplo, mesmo com um custo de frete mais alto, compensa essa despesa com um faturamento muito superior. A venda da Loja 4 permitirá ao Sr. João obter capital e concentrar os esforços de melhoria operacional nas lojas restantes, que já demonstram maior potencial de receita.

A venda da Loja 4 é a decisão mais estratégica, permitindo que o capital obtido com a transação seja reinvestido em um novo negócio com maior potencial.
Demais métricas foram desconsiderados por não apresentarem uma relevância no que diz respeito a venda estratégica da loja menos rentável.
